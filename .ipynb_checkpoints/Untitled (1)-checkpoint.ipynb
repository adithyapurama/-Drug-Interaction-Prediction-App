```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, f1_score, recall_score, precision_score, roc_curve, auc, precision_recall_curve
from keras.models import Sequential, load_model
from keras.layers import Dense, Dropout, BatchNormalization, Input
from keras.callbacks import EarlyStopping, ReduceLROnPlateau, ModelCheckpoint
from tensorflow.keras.regularizers import l1_l2
import snf
from tensorflow.keras.utils import to_categorical
```

```python
base_path = r"C:\Users\sprag\OneDrive\Attachments\Desktop\project"

# âœ… Load all datasets (unchanged)
df_chem_jaccard = pd.read_csv(f"{base_path}\\chem_Jacarrd_sim.csv", index_col=0)
df_atc = pd.read_csv(f"{base_path}\\ATCSimilarityMat.csv", index_col=0)
df_chemical = pd.read_csv(f"{base_path}\\chemicalSimilarityMat.csv", index_col=0)
df_target_jaccard = pd.read_csv(f"{base_path}\\target_Jacarrd_sim.csv", index_col=0)
df_pathway_jaccard = pd.read_csv(f"{base_path}\\pathway_Jacarrd_sim.csv", index_col=0)
df_enzyme_jaccard = pd.read_csv(f"{base_path}\\enzyme_Jacarrd_sim.csv", index_col=0)
df_sideeffect = pd.read_csv(f"{base_path}\\SideEffectSimilarityMat.csv", index_col=0)
df_offsideeffect = pd.read_csv(f"{base_path}\\offsideeffect_Jacarrd_sim.csv", index_col=0)
```

```python
matrices = [
    df_chem_jaccard.values,
    df_atc.values,
    df_chemical.values,
    df_target_jaccard.values,
    df_pathway_jaccard.values,
    df_enzyme_jaccard.values,
    df_sideeffect.values,
    df_offsideeffect.values
]
```

```python
# âœ… Find the most common shape
target_size = 548  # Change target size to 548

# âœ… Resize matrices to have consistent shape (548x548)
matrices = [np.pad(m, ((0, target_size - m.shape[0]), (0, target_size - m.shape[1])), mode='constant', constant_values=0) 
            if m.shape[0] < target_size or m.shape[1] < target_size else m[:target_size, :target_size] 
            for m in matrices]

print("\nâœ… Resized All Matrices to (548, 548)")

# âœ… Handle Zero Rows
for i in range(len(matrices)):
    row_sums = np.sum(matrices[i], axis=1, keepdims=True)
    zero_rows = (row_sums == 0).flatten()

    if np.any(zero_rows):
        print(f"âš  Matrix {i+1} had {np.sum(zero_rows)} zero rows. Fixing...")
        row_mean = np.mean(matrices[i][~zero_rows], axis=0)
        matrices[i][zero_rows] = row_mean

    matrices[i] /= np.maximum(row_sums, 1e-6)  # Normalize

# âœ… Apply SNF
fused_similarity = snf.snf(matrices, K=10, t=20)

print("\nâœ… Fused Similarity Matrix Created Successfully!")

# âœ… Save fused_similarity for Streamlit to use
np.save(f"{base_path}\\fused_similarity.npy", fused_similarity)
print(f"âœ… Saved fused_similarity to {base_path}\\fused_similarity.npy")
```

```python
# Load other datasets (unchanged)
df_txt = pd.read_csv(f"{base_path}\\IntegratedDS1.txt", sep=",", header=None)
df_drug_drug = pd.read_csv(f"{base_path}\\drug_drug_matrix.csv", index_col=0)
df_ddi = pd.read_csv(f"{base_path}\\filtered_ddiMatrix.csv", index_col=0)
df_drug_list = pd.read_csv(f"{base_path}\\drug_list.csv", sep=",", names=["Index", "DrugBank_ID", "Drug_Name"])

# âœ… Resize Drug-Drug Interaction Matrix (Ensure it matches 548x548)
interaction = df_drug_drug.values
target_size = 548

if interaction.shape[0] < target_size or interaction.shape[1] < target_size:
    interaction = np.pad(interaction, ((0, target_size - interaction.shape[0]), (0, target_size - interaction.shape[1])), 
                         mode='constant', constant_values=0)
else:
    interaction = interaction[:target_size, :target_size]

print(f"âœ… Resized Drug-Drug Interaction Matrix Shape: {interaction.shape}")

# âœ… Save df_drug_list for Streamlit to use
df_drug_list.to_csv(f"{base_path}\\drug_list_processed.csv", index=False)
print(f"âœ… Saved df_drug_list to {base_path}\\drug_list_processed.csv")
```

```python
datasets = {
    "IntegratedDS1": df_txt,
    "Drug-Drug Matrix": df_drug_drug,
    "DDI Matrix": df_ddi,
    "Chemical Jaccard": df_chem_jaccard,
    "ATC Similarity": df_atc,
    "Chemical Similarity": df_chemical,
    "Target Jaccard": df_target_jaccard,
    "Pathway Jaccard": df_pathway_jaccard,
    "Enzyme Jaccard": df_enzyme_jaccard,
    "Side Effect Similarity": df_sideeffect,
    "Offside Effect Jaccard": df_offsideeffect,
    "Drug List": df_drug_list
}
```

```python
def prepare_data():
    drug_fea = MinMaxScaler().fit_transform(datasets["IntegratedDS1"].values)
    
    # Ensure Drug-Drug Matrix is (548, 548)
    interaction = datasets["Drug-Drug Matrix"].values
    target_size = 548  # Adjust to 548

    if interaction.shape[0] < target_size or interaction.shape[1] < target_size:
        interaction = np.pad(interaction, ((0, target_size - interaction.shape[0]), (0, target_size - interaction.shape[1])), 
                             mode='constant', constant_values=0)
    else:
        interaction = interaction[:target_size, :target_size]

    train, label = [], []
    min_size = min(drug_fea.shape[0], fused_similarity.shape[0], interaction.shape[0])

    for i in range(min_size):
        for j in range(min_size):
            if i != j:
                combined_features = list(drug_fea[i]) + list(drug_fea[j]) + [fused_similarity[i, j]]
                train.append(combined_features)
                label.append(int(interaction[i, j]))

    train = np.array(train)
    label = np.array(label)
    label = to_categorical(label, num_classes=2)

    return train, label, drug_fea, fused_similarity

# âœ… Run the function
X, y, drug_fea, fused_similarity = prepare_data()
```

```python
# âœ… Save similarity matrices for Streamlit to use
similarity_matrices = [
    df_chem_jaccard.values,
    df_atc.values,
    df_chemical.values,
    df_target_jaccard.values,
    df_pathway_jaccard.values,
    df_enzyme_jaccard.values,
    df_sideeffect.values,
    df_offsideeffect.values
]
np.save(f"{base_path}\\similarity_matrices.npy", np.array(similarity_matrices, dtype=object))
print(f"âœ… Saved similarity_matrices to {base_path}\\similarity_matrices.npy")
```

```python
# Model definition and training (unchanged)
from tensorflow.keras.optimizers import Adam
def NDD(input_dim): 
    model = Sequential([
        Input(shape=(input_dim,)),
        BatchNormalization(),
        Dropout(0.4),
        Dense(512, activation='relu', kernel_regularizer=l1_l2()),  
        BatchNormalization(),
        Dropout(0.3),
        Dense(256, activation='relu', kernel_regularizer=l1_l2()),  
        BatchNormalization(),
        Dropout(0.3),
        Dense(128, activation='relu', kernel_regularizer=l1_l2()),  
        BatchNormalization(),
        Dropout(0.2),
        Dense(64, activation='relu', kernel_regularizer=l1_l2()),  
        BatchNormalization(),
        Dropout(0.2),
        Dense(2, activation='sigmoid')
    ])
    
    model.compile(loss='binary_crossentropy', optimizer=Adam(learning_rate=0.0005), metrics=['accuracy'])                  
    return model
```

```python
def train_model():
    print("ðŸ”„ Preparing Data...")
    X, y, _, _ = prepare_data()
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)

    print("ðŸš€ Initializing Model...")
    model = NDD(X_train.shape[1])

    early_stopping = EarlyStopping(monitor="val_loss", patience=10, restore_best_weights=True)
    lr_schedule = ReduceLROnPlateau(monitor="val_loss", factor=0.5, patience=5, min_lr=1e-6)
    model_checkpoint = ModelCheckpoint(f"{base_path}\\final_model.keras", monitor="val_loss", save_best_only=True)

    print("ðŸŸ¢ Training Started...")
    model.fit(
        X_train, y_train,
        batch_size=256, epochs=200, validation_split=0.1,
        callbacks=[early_stopping, lr_schedule, model_checkpoint]
    )

    print("âœ… Training Completed! Model Saved as 'final_model.keras'")

    # Evaluate model (unchanged)
    print("\nðŸ“Š Evaluating Model...")
    y_pred_prob = model.predict(X_test)
    y_pred = (y_pred_prob[:, 1] > 0.5).astype(int)
    y_test_labels = y_test[:, 1]

    accuracy = accuracy_score(y_test_labels, y_pred)
    precision = precision_score(y_test_labels, y_pred)
    recall = recall_score(y_test_labels, y_pred)
    f1 = f1_score(y_test_labels, y_pred)

    print(f"\nâœ… Evaluation Metrics:")
    print(f"ðŸ”¹ Accuracy:  {accuracy:.4f}")
    print(f"ðŸ”¹ Precision: {precision:.4f}")
    print(f"ðŸ”¹ Recall:    {recall:.4f}")
    print(f"ðŸ”¹ F1 Score:  {f1:.4f}")

# Run training
train_model()
```

```python
# âœ… Import required libraries
import numpy as np

# âœ… Prediction function
def predict_interaction(drug1_name, drug2_name, drug_fea, model, similarity_matrices, df_drug_list, drug_info_dict):
    
    # âœ… Get the list of drug names
    drug_names = list(df_drug_list["Drug_Name"])
    
    # âœ… Check if both drugs exist in the dataset
    if drug1_name not in drug_names or drug2_name not in drug_names:
        return {"Error": "One or both drugs not found!"}

    # âœ… Get indices of the drugs
    idx1 = drug_names.index(drug1_name)
    idx2 = drug_names.index(drug2_name)

    # âœ… Extract drug features
    drug1_features = drug_fea[idx1]
    drug2_features = drug_fea[idx2]

    # âœ… Extract similarity scores
    similarities = [sim[idx1, idx2] for sim in similarity_matrices]

    # âœ… Prepare input for model
    expected_input_size = model.input_shape[1]
    combined_features = np.concatenate([drug1_features.flatten(), drug2_features.flatten(), similarities])

    # âœ… Adjust feature size to match model input
    if combined_features.shape[0] > expected_input_size:
        combined_features = combined_features[:expected_input_size]
    else:
        combined_features = np.pad(combined_features, (0, expected_input_size - combined_features.shape[0]))

    # âœ… Reshape input and make prediction
    combined_features = combined_features.reshape(1, -1)
    prediction = model.predict(combined_features)

    # âœ… Process prediction result
    if prediction.shape[-1] == 2:
        predicted_label = np.argmax(prediction)
        probability = float(prediction[0][1])
    else:
        predicted_label = int(prediction[0] > 0.5)
        probability = float(prediction[0])

    # âœ… Retrieve drug IDs safely
    drug1_id_values = df_drug_list.loc[df_drug_list["Drug_Name"] == drug1_name, "DrugBank_ID"].values
    drug2_id_values = df_drug_list.loc[df_drug_list["Drug_Name"] == drug2_name, "DrugBank_ID"].values

    drug1_id = drug1_id_values[0] if len(drug1_id_values) > 0 else "N/A"
    drug2_id = drug2_id_values[0] if len(drug2_id_values) > 0 else "N/A"

    # âœ… Debugging: Print available drug IDs
    print(f"Drug 1 ID: {drug1_id}, Drug 2 ID: {drug2_id}")
    print(f"Available Drug IDs in drug_info_dict: {list(drug_info_dict.keys())[:10]}")  # Print first 10 keys

    # âœ… Retrieve additional drug information safely
    drug1_info = drug_info_dict.get(drug1_name, {"Side Effects": "N/A", "Target Sites": "N/A"})
    drug2_info = drug_info_dict.get(drug2_name, {"Side Effects": "N/A", "Target Sites": "N/A"})

    # âœ… Return final result
    return {
        "Drug 1": drug1_name,
        "Drug 2": drug2_name,
        "Interaction": "Yes" if predicted_label == 1 else "No",
        "Probability": probability,
        "Drug 1 Side Effects": drug1_info["Side Effects"],
        "Drug 2 Side Effects": drug2_info["Side Effects"],
        "Drug 1 Target Sites": drug1_info["Target Sites"],
        "Drug 2 Target Sites": drug2_info["Target Sites"]
    }
```

```python
import numpy as np

def predict_interaction(drug1_name, drug2_name, drug_fea, model, similarity_matrices, df_drug_list, drug_info_dict):
    drug_names = list(df_drug_list["Drug_Name"])

    # âœ… Check if both drugs exist in the dataset
    if drug1_name not in drug_names or drug2_name not in drug_names:
        return {"Error": "One or both drugs not found!"}

    # âœ… Get indices of the drugs
    idx1 = drug_names.index(drug1_name)
    idx2 = drug_names.index(drug2_name)

    # âœ… Extract drug features
    drug1_features = drug_fea[idx1]
    drug2_features = drug_fea[idx2]

    # âœ… Extract similarity scores
    similarities = [sim[idx1, idx2] for sim in similarity_matrices]

    # âœ… Prepare input for model
    expected_input_size = model.input_shape[1]
    
    def prepare_input(features1, features2, sims):
        combined_features = np.concatenate([features1.flatten(), features2.flatten(), sims])
        if combined_features.shape[0] > expected_input_size:
            combined_features = combined_features[:expected_input_size]
        else:
            combined_features = np.pad(combined_features, (0, expected_input_size - combined_features.shape[0]))
        return combined_features.reshape(1, -1)

    # âœ… Predict both orders
    input1 = prepare_input(drug1_features, drug2_features, similarities)
    input2 = prepare_input(drug2_features, drug1_features, similarities)

    prediction1 = model.predict(input1)
    prediction2 = model.predict(input2)

    def process_prediction(prediction):
        if prediction.shape[-1] == 2:
            return np.argmax(prediction), float(prediction[0][1])
        return int(prediction[0] > 0.5), float(prediction[0])

    predicted_label1, probability1 = process_prediction(prediction1)
    predicted_label2, probability2 = process_prediction(prediction2)

    # âœ… Compute final averaged probability
    final_probability = (probability1 + probability2) / 2
    final_label = "Yes" if (predicted_label1 == 1 or predicted_label2 == 1) else "No"

    # âœ… Retrieve drug IDs safely
    drug1_id_values = df_drug_list.loc[df_drug_list["Drug_Name"] == drug1_name, "DrugBank_ID"].values
    drug2_id_values = df_drug_list.loc[df_drug_list["Drug_Name"] == drug2_name, "DrugBank_ID"].values

    drug1_id = drug1_id_values[0] if len(drug1_id_values) > 0 else "N/A"
    drug2_id = drug2_id_values[0] if len(drug2_id_values) > 0 else "N/A"

    print(f"Drug 1 ID: {drug1_id}, Drug 2 ID: {drug2_id}")
    print(f"Available Drug IDs in drug_info_dict: {list(drug_info_dict.keys())[:10]}")  

    # âœ… Retrieve additional drug information safely
    drug1_info = drug_info_dict.get(drug1_name, {"Side Effects": "N/A", "Target Sites": "N/A"})
    drug2_info = drug_info_dict.get(drug2_name, {"Side Effects": "N/A", "Target Sites": "N/A"})

    return {
        "Drug 1": drug1_name,
        "Drug 2": drug2_name,
        "Interaction": final_label,
        "Probability": final_probability,
        "Drug 1 Side Effects": drug1_info["Side Effects"],
        "Drug 2 Side Effects": drug2_info["Side Effects"],
        "Drug 1 Target Sites": drug1_info["Target Sites"],
        "Drug 2 Target Sites": drug2_info["Target Sites"]
    }
```
